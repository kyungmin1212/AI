# Multi-variable linear regression

```python
import tensorflow as tf
import numpy as np
```

```python
x_data=[[73.,80.,75.],[93.,88.,93.],[89.,91.,90.],[96.,98.,100.],[73.,66.,70.]]
y_data=[[152.],[158.],[180.],[196.],[142.]]
```

```python
tf.model=tf.keras.Sequential()
```

### tf.model.add(tf.keras.layers.Activation('linear')) 은 생략가능 선형 Activation이 기본값
### tf.model.add(tf.keras.layers.Dense(units=1,input_dim=3),activation='linear) 한번에 쓰는것 가능

```python
tf.model.add(tf.keras.layers.Dense(units=1,input_dim=3))
tf.model.add(tf.keras.layers.Activation('linear'))
```

```python
tf.model.compile(loss='mse',optimizer=tf.keras.optimizers.SGD(lr=1e-5))
```

```python
tf.model.summary()
```

```python
history=tf.model.fit(x_data,y_data,epochs=100)
```

```python
y_predict=tf.model.predict(np.array([[72.,93.,96.]]))
print(y_predict)
```
