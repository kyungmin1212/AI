## Tensor = data array
## Tensor Ranks = 몇차원데이터
## Tensor Shapes = 차원데이터개수표현
## Tensor Types = 정수,실수 등

```python
v=tf.constant([1,2,3,4])
m=tf.constant([[1,2,3],[4,5,6],[7,8,9]])
t=tf.constant([[[2,4],[5,6]],[[3,1],[1,2]]])

print(tf.rank(v))
print(tf.shape(v))
print(tf.rank(m))
print(tf.shape(m))
print(tf.rank(t))
print(tf.shape(t))
```
